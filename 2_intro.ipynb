```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#how to read a jpg or png file with cv2 library
img=cv2.imread('sample_data/00-puppy.jpg')
```

```python
type(img)
#if it says none type, this means that the path is not correct
```

```python
# The 3rd dimension comes from the colors(The RGB channel)
img.shape
```

```python
plt.imshow(img)
```

```python
#openCV --> Blue, Green, Red
#matplot --> Red, Green, Blue

fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
fix_img.shape
```

```python
#flipping the image around the x-axis
flip_img = cv2.flip(fix_img, 0)
```

```python
plt.imshow(flip_img)
```

```python
#flipping the image around y-axis
flip2_img=cv2.flip(fix_img,1)
plt.imshow(flip2_img)
```

```python
#flipping the image around x and y-axis
flip3_img=cv2.flip(fix_img,-1)
plt.imshow(flip3_img)
```

```python
gray = cv2.cvtColor(fix_img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
small_img=cv2.resize(fix_img,(500,400)) #(width,height)
plt.imshow(small_img)
```

```python
small_img.shape
```

```python
ratio_img=cv2.resize(fix_img,(0,0),fix_img,0.5,0.5)
```

```python
plt.imshow(ratio_img)
```
